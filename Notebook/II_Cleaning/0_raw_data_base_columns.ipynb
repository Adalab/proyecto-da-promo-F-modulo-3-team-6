Age

```python
#!pip install word2number
```

```python
#Para transformar números (tanto letras como números en strings) a enteros
from word2number import w2n
import numpy as np
```

```python
#Función que transforma las palabras (en inglés) y strings de números ('54') a integers
def transform_to_number(age):
    #Hemos puesto try y except porque al hacerlo una vez funcionaba una la segunda vez daba error
    try:
        return w2n.word_to_num(age)
    #Con este except le decimos que devuelva el integer sin cambiarlo
    except:
        return age 

#El método .apply en pandas se utiliza para aplicar una función a lo largo de un eje específico de un DataFrame
#Pasa por cada uno de los elementos de la columna
df_rawdata["Age"] = df_rawdata["Age"].apply(transform_to_number)

#Imprimimos la lista para comprobar
print(df_rawdata["Age"].tolist())
```

```python
print("La edad mínima es de: ", np.min(df_rawdata["Age"]), "años y la edad máxima es de:", np.max(df_rawdata["Age"])," años")
print("La media de edad es de", df_rawdata["Age"].mean())
```
