- Variable Daily Rate: 

```python
#Función que limpia los números de salarios (cambia , por . y quita simbolo $)
def transform_salary(wage):
    try:
        #Con replace quitamos 1º el símbolo del dolar, 2º las comas por puntos y quitamos espacios extra
        return float(wage.replace("$","").replace(",",".").strip())
    except:
        # Si los datos ya han sido cambiados nos los devuelve igual.
        return wage

#El método .apply en pandas se utiliza para aplicar una función a lo largo de un eje específico de un DataFrame
#Pasa por cada uno de los elementos de la columna sobreescribiendo el nuevo valor
df_rawdata["Dailyrate"] = df_rawdata["Dailyrate"].apply(transform_salary)
```

        - CAMBIAR LABELS de las variables "satisfaction" que son de 1-4 a "muy satisf-nada satisf"

```python
# Jobsatisfaction
# Relationshipsatisfaction
# Environmentsatisfaction

# 1 - Not Satisfied
# 2 - Somewhat Satisfied
# 3 - Satisfied
# 4 - Very Satisfied
```

```python
df_rawdata.head()
```

FUNCIONA :D 

```python
def labels_satisfaction(level):
    dict_map = {1: "Not Satisfied", 2: "Somewhat Satisfied", 3: "Satisfied", 4: "Very Satisfied"}
    
    try: 
        df_rawdata[level] = df_rawdata[level].map(dict_map)
        #df_rawdata["Jobsatisfaction"].head(5)
        return df_rawdata[level]
    except: 
        return level 
    
columns_to_modify = ["Jobsatisfaction", "Relationshipsatisfaction", "Environmentsatisfaction"]
for column in columns_to_modify:
    df_rawdata[column] = labels_satisfaction(column)
```

```python
df_rawdata["Environmentsatisfaction"].head()
```

        - Cambiando la variable job level, de 1 a 5, por labels

```python
df_rawdata["Joblevel"].value_counts()
```

```python
def labels_joblevel(level):
    dict_job = {
1: "Entry-level",
2: "Junior",
3: "Mid-level",
4: "Senior",
5: "Executive"
}
    
    try: 
        df_rawdata[level] = df_rawdata[level].map(dict_job)
        #df_rawdata["Jobsatisfaction"].head(5)
        return df_rawdata[level]
    except: 
        return level 
    
columns_to_modify = ["Joblevel"]
for column in columns_to_modify:
    df_rawdata[column] = labels_joblevel(column)
```

```python
df_rawdata["Joblevel"].value_counts()
```

        - variable Marital Status: replace errores de texto

```python
df_rawdata['Maritalstatus'] = df_rawdata['Maritalstatus'].str.title()
```

```python
df_rawdata['Maritalstatus'] = df_rawdata['Maritalstatus'].replace('Marreid', "Married")
```

```python
df_rawdata["Maritalstatus"].value_counts()
```

```python
def cambiar_comas(cadena):
    try:
        return float(cadena.replace(",", "."))
    except:
        return np.nan
    
df_rawdata["Worklifebalance"] = df_rawdata["Worklifebalance"].apply(cambiar_comas)
```
