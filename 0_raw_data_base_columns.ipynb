 - Vamos a transformar las columnas del DF en capitalize para unificar criterios

```python
df_rawdata.columns
```

- For loop para cada una de las columnas. Guardamos en una variable nueva "new_columns"

```python
new_columns={columna: columna.capitalize() for columna in df_rawdata.columns} 
new_columns
```

- Renombramos las columnas con los valores de new_columns en el mismo DF con el que venimos trabajando

```python
df_rawdata.rename(columns = new_columns, inplace = True)
df_rawdata.head()
```
